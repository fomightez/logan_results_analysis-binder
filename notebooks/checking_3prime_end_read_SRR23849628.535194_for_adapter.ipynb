## checking 3' end read_SRR23849628.535194 for adapter

Is that adapter sequences that is seen after what looks to correspond to the poly-A tail in read @SRR23849628.535194 from `USP7_locus_mapped_SRX19662564_reads.fastq`?   
Just a quick assessment to see if it, or somewhat different versions given nanopore reads can be error prone, occur often in the reads.

This notebook run in sessions launched from blast-binder because it has PatMatch installed in sessions, could also have used patmatch-binder sessions. ( [my blast-binder repo](https://github.com/fomightez/blast-binder) or [my patmatch-binder repo](https://github.com/fomightez/patmatch-binder))

```python
%%bash
awk '{if(NR%4==1) {printf(">%s\n",substr($0,2));} else if(NR%4==2) print;}' USP7_locus_mapped_SRX19662564_reads.fastq > output.fasta
```

```python
!curl -O https://raw.githubusercontent.com/fomightez/sequencework/master/patmatch-utilities/patmatch_results_to_df.py
```

```python
%%capture
adapter_candidate = 'CGTTGATCGGAAGAGCGTCGTGTAGAGCAATACGTAACTGAA'
#!fqgrep --reverse-complement --color auto {USP7_TSL4_RET_I} USP7_locus_mapped_SRX19662564_reads.fastq

pm_result = !perl patmatch_1.2/patmatch.pl -c {adapter_candidate} output.fasta 6 ids
from patmatch_results_to_df import patmatch_results_to_df
my_pattern= "CGTTGATCGGAAGAGCGTCGTGTAGAGCAATACGTAACTGAA"
df = patmatch_results_to_df(pm_result.n, pattern=my_pattern, name="3prime_end_of_read")
```

I could explore this more by looking further at position of each pattern match in the sequence, but the first one I looked at , other than the original from `@SRR23849628.535194`, using 'find' in a text editor `GTAGATTGGAAGAGCGTCGTGTAGAGCAATACGTAACTGAA` from `SRR23849628.1017513` showed it was in four different ones at the end. This, I suspect the trend holds. Seems sufficient to support it being an adapter sequence of some sort.
